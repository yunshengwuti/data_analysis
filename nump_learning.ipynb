```python
import numpy as np
```

```python
arr = np.array([[1, 2, 3],[5, 6, 7]])
print(arr)
print(arr.ndim)      #数组维度
print(arr.shape)     #数组形状
print(arr.size)      #数组元素个数
print(arr.dtype)     #数组数据类型
print(arr.T)         #数组的转置
```

```python
arr = np.array([1, 2, 3],dtype=np.float64)
print(arr)
```

```python
arr1 = np.copy(arr)
print(arr1)
```

```python
arr = np.zeros((2,3),dtype=int)
print(arr)
```

```python
arr = np.zeros((20,),dtype=int)
print(arr)
```

```python
arr = np.ones((3, 3),dtype=int)
print(arr)
```

```python
arr = np.empty((3, 3))
```

```python
arr = np.full((3, 3), 2026)
print(arr)
```

```python
arr = np.zeros_like(arr)
print(arr)
```

```python
arr = np.arange(2, 11, 2)  # start, stop, step
print(arr) # 不包括stop
```

```python
arr = np.linspace(0,100,5,dtype=int) # 取多少个数
print(arr)    # 包括stop
arr = np.arange(0,101,25)
print(arr)
```

```python
arr = np.logspace(0,4,3,base=2)
print(arr)
```

```python
arr = np.eye(3,dtype=int) # 单位矩阵
print(arr)
arr = np.diag([5,1,2,3])  # 对角矩阵
print(arr)
```

```python
# 随机浮点数(均匀分布)
arr = np.random.rand(2,3)  # 随机数组2*3
print(arr)
arr = np.random.uniform(3,6,(2,3))  #指定范围随机数
print(arr)
```

```python
# 随机整数（均匀分布）
arr = np.random.randint(3, 30,(2,3))
print(arr)
```

```python
# 随机数列（正态分布）
arr = np.random.randn(2,3)
print(arr)
```

```python
# 设置随机种子
np.random.seed(20)
arr = np.random.randint(1,10,(2,5))
print(arr)
```
