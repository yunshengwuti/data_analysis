```python
import pandas as pd
df = pd.read_csv('20260611_信用_经纪权限.csv',encoding='gbk')
df.head()
```

```python
cols = ['国利','国际','中诚','平安','信唐','上田']
df['已开通数量'] = df[cols].apply(lambda row: row.str.contains('已开通').sum(), axis=1)
df.head
```

```python
df.sort_values('已开通数量', ascending=False).head(10)
```

```python
df.groupby('机构')['已开通数量'].mean().sort_values(ascending=False).head(10)
```

```python
df.groupby('机构')['已开通数量'].agg(['count','mean'])
```

```python
df_caculate = df.groupby('机构').agg(人数 = ('姓名', 'count'), 平均开通数量 = ('已开通数量','mean'))
```

```python
df_caculate[df_caculate['人数'] > 10].sort_values('人数',ascending=False)
```

```python
result = df_caculate[df_caculate['人数'] > 10].sort_values('平均开通数量',ascending=False)
result.to_excel('机构统计.xlsx',index=True)
```
